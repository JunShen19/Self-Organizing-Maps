```python
df = pd.read_csv(r'D:\JunShen\SOM\result\result_HQ_九隆(3)_LSTM_set6.csv', index_col=0)
```

```python
pred = np.array(())
true = np.array(())
for col in df.columns:
    if 'pred' in col:
        pred = np.append(pred, df.loc[:,col].values)
    else:
        true = np.append(true, df.loc[:,col].values)
```

```python
well_name = '九隆(3)'
font = {'weight' : 'normal',
    'size'   : 22}

plt.rc('font', **font)
```

```python
fig, ax = plt.subplots(figsize=(20, 16), dpi=80)

ax.set_xlabel("DATE")
ax.set_ylabel("GroundWater Level")
ax.set_title(f"{well_name}",  fontproperties="SimSun")

ax.plot(pred, color = 'r', label='predction')
ax.plot(true, color = 'g', label='Measurements')

fig.show()
```
